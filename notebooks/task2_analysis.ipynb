```python

import pandas as pd


import matplotlib.pyplot as plt


%matplotlib inline
```

```python
# Load stock price dataset
df = pd.read_csv("../data/raw/AAPL.csv")

# Display first 5 rows
df.head()
```

```python
# Show basic information about the dataset
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Check data types again
df.info()
```

```python
# Sort dataset by date
df = df.sort_values(by='Date')

# Display first 5 rows
df.head()
```

```python
# Create figure size
plt.figure(figsize=(12,6))

# Plot closing prices
plt.plot(df['Date'], df['Close'])

# Add title
plt.title('Closing Price Over Time')

# Label x-axis
plt.xlabel('Date')

# Label y-axis
plt.ylabel('Closing Price')

# Rotate dates for readability
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
# Calculate 20-day Simple Moving Average
df['SMA_20'] = df['Close'].rolling(window=20).mean()

# Calculate 50-day Simple Moving Average
df['SMA_50'] = df['Close'].rolling(window=50).mean()

# Display first rows
df.head()
```

```python
# Create figure
plt.figure(figsize=(14,7))

# Plot closing price
plt.plot(df['Date'], df['Close'], label='Close Price')

# Plot 20-day SMA
plt.plot(df['Date'], df['SMA_20'], label='20-Day SMA')

# Plot 50-day SMA
plt.plot(df['Date'], df['SMA_50'], label='50-Day SMA')

# Add title
plt.title('Closing Price with SMA Indicators')

# Label axes
plt.xlabel('Date')
plt.ylabel('Price')

# Rotate dates
plt.xticks(rotation=45)

# Show legend
plt.legend()

# Display plot
plt.show()
```
